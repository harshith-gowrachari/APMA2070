```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.fftpack import fft, ifft, fftfreq

def fourier_fprime(f, N):
    k=np.zeros(f.size)
    if ((N%2)==0):
        for i in range(1,N//2):
            k[i]=i
            k[N-i]=-i
    else:
        for i in range(1,(N-1)//2):
            k[i]=i
            k[N-i]=-i            
    fprime_fourier = ifft(1j*k*fft(f))
    return k, fprime_fourier

    
```

```python
N=100
x=2*np.pi*np.arange(0, N, 1)/N
dx = x[1]-x[0]
a = 1
b = 2
f = lambda x, a, b: np.sin(a*x) + np.cos(b*x)
f_prime_act = lambda x, a, b: a*np.cos(a*x) - b*np.sin(b*x)
plt.plot(x, f(x, a, b), "-g", label="f(x)")
plt.plot(x, f_prime_act(x, a, b), "-b", label="f'(x)")
plt.legend()
```

```python
y = f(x, a, b)
y_prime_analy = f_prime_act(x, a, b)
k, df_fourier = fourier_fprime(y, N)
point_err = y_prime_analy - df_fourier
L2_err = np.linalg.norm(y_prime_analy - df_fourier)/np.linalg.norm(y_prime_analy)
print(f"L2 Error is: {L2_err}")
```

```python
plt.plot(x, f_prime_act(x, a, b),"-k", lw=2.5, label="f'(x) Analytical")
plt.plot(x, df_fourier, "--g",lw=1.5, label="f'(x) Fourier")
plt.plot(x, point_err, "--b",lw=1.5, label="Error")

plt.legend()
```
